```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
%%time
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
